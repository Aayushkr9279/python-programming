```python
while True:
    num = float(input("Enter a number: "))

    if num > 0:
        print("The number is Positive.")
    elif num < 0:
        print("The number is Negative.")
    else:
        print("The number is Zero.")

    again = input("Do you want to check another number? (yes/no): ").lower()

    if again != "yes":
        print("Goodbye")
        break
```
