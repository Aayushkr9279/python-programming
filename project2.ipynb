```python
correct_pin = "1234"
balance = 5000
attempts = 3            
while attempts > 0:
    pin = input("Enter your 4-digit PIN: ")

    if pin == correct_pin:
        print("PIN accepted")
        print("Your balance is:",balance)
        break
    else:
        attempts -= 1
        print(f"Incorrect PIN! Attempts left: {attempts}")

        if attempts == 0:
            print("Your card has been blocked due to too many incorrect attempts.")
```
